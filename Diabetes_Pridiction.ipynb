```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("iammustafatz/diabetes-prediction-dataset")

print("Path to dataset files:", path)
```

```python
# The path where the dataset was downloaded by kagglehub
dataset_path = '/root/.cache/kagglehub/datasets/iammustafatz/diabetes-prediction-dataset/versions/1'

# Assuming the CSV file is named 'diabetes_prediction_dataset.csv'
csv_file_path = f'{dataset_path}/diabetes_prediction_dataset.csv'

# Load the data into a new pandas DataFrame
kaggle_diabetes_data = pd.read_csv(csv_file_path)

# Display the first few rows to confirm
print("First 5 rows of the Kaggle dataset:")
display(kaggle_diabetes_data.head())

# Display the shape of the new DataFrame
print("\nShape of the Kaggle dataset DataFrame:")
print(kaggle_diabetes_data.shape)
```

```python
kaggle_diabetes_data.shape
```

```python
kaggle_diabetes_data['diabetes'].value_counts()
```

```python
numerical_columns=kaggle_diabetes_data.select_dtypes(include=np.number)
```

```python
numerical_columns.groupby('diabetes').mean()
```

Seprating data and labels

```python
X = numerical_columns.drop(columns='diabetes',axis=1)
Y = numerical_columns['diabetes']
```

```python
print(X)
print(Y)
```

Data Standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
Standardized_data=scaler.transform(X)
```

```python
print (Standardized_data)
```

```python
X =Standardized_data
Y = numerical_columns['diabetes']
```

Train Test Split

```python
X_test,X_train,Y_test,Y_train=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

Model evalution

```python
X_train_prediction=classifier.predict(X_train)
```

```python
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuracy)
```
